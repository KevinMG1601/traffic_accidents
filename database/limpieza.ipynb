```python
import pandas as pd
import numpy as np
```

```python
csv = ("../data/df.csv")
df = pd.read_csv(csv, delimiter=',')
print(df.describe)
```

```python
#remplazamos los campos vacios por null
df = df.replace(r'^\s*$', np.nan, regex=True)

#sacamos el a√±o de la fecha crash_date
df["crash_year"] = pd.to_datetime(df["crash_date"], format="%m/%d/%Y %I:%M:%S %p").dt.year

#eliminamos las columnas que no necesitamos
columnas_eliminar = ["alignment", "road_defect", "num_units", "crash_date", "damage"]
df = df.drop(columns=[col for col in columnas_eliminar if col in df.columns])

print(df.describe)
```

```python
df.to_csv("../data/df_limpio.csv", index=False, na_rep="NULL")
```
